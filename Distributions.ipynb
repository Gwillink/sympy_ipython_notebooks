```python
x = Symbol("x")
```

```python
def plot_dist(rv, paramdict, urange):
    D = Density(rv)
    f = D[1].subs(paramdict)
    # Fails for piecewise objects!
    #F = lambdify([D[0]], f, "numpy")
    
    u = np.linspace(urange[0], urange[1], 500)
    #v = F(u)
    v = np.array([f.subs(D[0],i).evalf() for i in u], dtype=np.float)
    
    m = np.amax(v)
    
    col = list(np.random.rand(3))
    
    plot.figure()
    plot.fill_between(u,np.zeros(u.shape),v, color=col+[0.5])
    plot.plot(u,v, color=col)
    plot.grid(True)
    plot.xlim(urange)
    plot.ylim([0.0, 1.1*m])
```

```python
def Mode(X):
    """The mode is defined as arg max(density(X))
    """
    D = Density(X)
    df = diff(D[1], D[0])
    mode = solve(Eq(df,0), D[0])
    if len(mode) == 1:
        return mode[0]
    return mode
```

```python
def MGF(X, symbol=None):
    if not symbol:
        symbol = Dummy("t")
    t = symbol
    
    mgf = E(exp(t*X), meijerg=True)
    
    return Tuple(t, mgf)
```

```python
def Moment(X, k=1):
    mu = E(X, meijerg=True)
    
    kmom = E((X-mu)**k)
    
    return kmom
```

```python
def Kurtosis(X):
    mu4 = Moment(X, 4)
    va4 = Var(X, meijerg=True)
    
    k = simplify(mu4 / va4**2 - 3)
    
    return k
```
