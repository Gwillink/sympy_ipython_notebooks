```python
from sympy import *
from sympy.stats import *
```

```python
import numpy as np
import matplotlib.pyplot as plot
```

```python
x = Symbol("x")
```

```python
def plot_dist(rv, paramdict, urange):
    D = Density(rv)
    f = D[1].subs(paramdict)
    # Fails for piecewise objects!
    #F = lambdify([D[0]], f, "numpy")
    
    u = np.linspace(urange[0], urange[1], 500)
    #v = F(u)
    v = np.array([f.subs(D[0],i).evalf() for i in u], dtype=np.float)
    
    m = np.amax(v)
    
    col = list(np.random.rand(3))
    
    plot.figure()
    plot.fill_between(u,np.zeros(u.shape),v, color=col+[0.5])
    plot.plot(u,v, color=col)
    plot.grid(True)
    plot.xlim(urange)
    plot.ylim([0.0, 1.1*m])
```

See also: http://en.wikipedia.org/wiki/List_of_probability_distributions
