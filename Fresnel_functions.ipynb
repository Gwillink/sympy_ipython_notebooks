Plot of the two Fresnel integrals $S(x)$ and $C(x)$

```python
t = linspace(0, 8, 8000)
u = [ fresnels(i).evalf() for i in t ]
v = [ fresnelc(i).evalf() for i in t ]
u = array(u)
v = array(v)
```

```python
figure(figsize=(10,8))
plot(t, u, label=r"$S(x)$")
plot(t, v, label=r"$C(x)$")
grid(True)
legend(loc="best")
```

The Cornu spiral defined as the parametric curve $u(t),v(t) := C(t), S(t)$

```python
figure(figsize=(8,8))
plot(v, u, "b")
plot(-v, -u, "b")
grid(True)
```
