Examples of sine transformations. We use the definition
$$
\mathcal{SIN}_t \left[f(t)\right]\left(w\right) := \frac{\sqrt{2}}{\sqrt{\pi}} \int_0^{\infty} f(t) \sin(t \cdot w) \mathrm{d}t
$$

```python
t = Symbol("t")
w = Symbol("w")
```

```python
n = Symbol("n", integer=True)
```

```python
a = Symbol("a")
b = Symbol("b")
```
